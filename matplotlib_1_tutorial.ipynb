```python
#real data 

#oscar Winner Movies :
oscar_movies = [
    "The Dark Knight",
    "The Hurt Locker",
    "The King's Speech",
    "The Artist",
    "Argo"
]
years =[2008,2009,2010,2011,2012]
oscar_revenue=[1005,170,427,133,132]


plt.plot(years,oscar_revenue)

plt.title("Dscar Movies Revenue in each year")
plt.xlabel("Years")
plt.ylabel("Revenue(in $M)")
```

```python
#non oscar Winner Movies 
non_oscar_movies =["Slumdog Milliniore" , "Avatar", "Inception", "Hugo", "Licoln"]
non_oscar_revenue = [378,2788,829,185 ,275]


plt.plot(years,oscar_revenue,color="green",marker="<",linestyle="--",linewidth=3 ,label="Oscar Movies")
plt.plot(years,non_oscar_revenue ,label="Non-Oscar Movies")

plt.title("Oscar Movies vs Non-Oscar Movies Revenue")
plt.xlabel("Year")
plt.ylabel("Revenue (in $M)")

plt.legend()
```
